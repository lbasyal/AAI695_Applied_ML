```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy.polynomial.polynomial as poly
```

```python
noise_scale= 100
number_of_samples= 50
x =  25*(np.random.rand(number_of_samples, 1) -0.5)
y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

## 1. Please plot the noisy data and the polynomial you found (in the same figure). You can use any value of m selected from 2, 3, 4, 5, 6.

```python
fitted=[]
for i in range(2,7):
    z=np.polyfit(x.flatten(), y.flatten(), i)
    p=np.poly1d(z)
    fitted.append(p(x.flatten()))
```

```python
plt.plot(x,y,'ro')
plt.plot(x.flatten(),fitted[0].flatten(),'bo')
plt.plot(x.flatten(),fitted[1].flatten(),'go')
plt.plot(x.flatten(),fitted[2].flatten(),'mo')
plt.plot(x.flatten(),fitted[3].flatten(),'co')
plt.plot(x.flatten(),fitted[4].flatten(),'ko')
plt.legend()
```

## 2. Plot MSE versus order m,for m = 1, 2, 3, 4, 5, 6, 7, 8 respectively.Identify the best choice of m.

```python
#Get plot for m values
mse= []
for i in range(1,9):
    z=np.polyfit(x.flatten(), y.flatten(), i)
    p=np.poly1d(z)
    out=(p(x.flatten()))
    mse.append((np.square(out.flatten() - y.flatten())).mean(axis=0))
xx=np.array((1,2,3,4,5,6,7,8))    
```

```python
plt.plot(xx,mse,'r-')
```
